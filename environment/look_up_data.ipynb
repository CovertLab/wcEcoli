```python
# import matplotlib, pandas, numpy, csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv

# make matplotlib output appear inline
%matplotlib inline
plt.rcParams['figure.figsize'] = [14, 5]

# import test_data.csv from surrogate experiments as pandas dataframe 'td'
td = pd.read_csv('test_data.csv')
# display the header (first 5 lines) of td to check successful import
td.head()

# generate dictionary of unique agent_ids and their index values 
agent_master = td.groupby('agent_id').groups
for index in agent_master:
    print('agent_id = ' + str(index)) + ', index = ' + str(agent_master[index][0])

    # generate dictionary of unique agents for plotting
agent_plot = {}
i = 0
for index in agent_master:
    agent_plot[i] = td[td.agent_id == td.agent_id[agent_master[index][0]]]
    agent_plot[i] = agent_plot[i].drop(agent_plot[i].index[0])
    i+=1
```

```python
# compare agents' volume & internal glucose counts
time = agent_plot[0]['timestep']
plt.figure(1)
plt.subplot(121)
for index in agent_plot:
    plt.plot(agent_plot[index]['timestep'], agent_plot[index]['GLT[c]'])
plt.title("[GLC]")

plt.figure(1)
plt.subplot(122)
for index in agent_plot:
    plt.plot(agent_plot[index]['timestep'], agent_plot[index]['volume'])
plt.title("Volume")
```
