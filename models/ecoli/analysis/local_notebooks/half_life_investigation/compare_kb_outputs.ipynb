```python
def check_column_differences(df1, df2):
    col_names = set(df1.columns) | set(df2.columns)
    print(f'Columns in df1: {set(df1.columns)} ({len(set(df1.columns))} columns)')
    print(f'Columns in df2: {set(df2.columns)} ({len(set(df2.columns))} columns)')
    print(f'All columns: {col_names} ({len(col_names)} columns)')
    protein_ids = set(df1['monomer_id']) | set(df2['monomer_id'])
    for col in col_names:
        # check if the contents of the column are the same in both dataframes:
        if col in df1.columns and col in df2.columns:
            for monomer in protein_ids:
                val1 = df1[df1['monomer_id'] == monomer][col].values
                val2 = df2[df2['monomer_id'] == monomer][col].values
                if not np.array_equal(val1, val2):
                    print(f'Column "{col}" has different values for monomer "{monomer}" in df1 and df2.')
                    print(f'Value in df1: {val1}')
                    print(f'Value in df2: {val2}')
            
                
                
```

```python
check_column_differences(df1, df2)
```
